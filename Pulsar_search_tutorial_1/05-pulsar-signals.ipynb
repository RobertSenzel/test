# Pulsar signals

Pulsars emit radio beams from both their **north and south magnetic poles**.

As the pulsar spins, these beams sweep across space.  
When one of the beams points toward Earth, we detect a **pulse** of radio waves.

This is why pulsars appear to “pulse”.

### What does this look like in our data?

In a time series, a pulsar signal appears as **regular, repeating spikes**.

- When the beam is **not** pointing toward Earth, we detect very few (or no) photons  
- When the beam **sweeps across Earth**, we see a sudden increase in radio signal

Let's plot what this looks like.

```python
import numpy as np
import matplotlib.pyplot as plt

time, psr_slow = np.load("data/pulsar_signal_slow.npy")
time, psr_fast = np.load("data/pulsar_signal_fast.npy")
```

```python
# slow pulsar
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(time, psr_slow, ls='-', lw=0.5, marker='o', markersize=2)
ax.set_title('Slow pulsar')
ax.set_xlabel('Observation time (s)', fontsize=12)
ax.set_ylabel('Photon count', fontsize=12)
```

```python
# fast pulsar
fig, ax = plt.subplots(figsize=(20, 6))
ax.set_title('Fast pulsar')
ax.plot(time, psr_fast, ls='-', lw=0.5, marker='o', markersize=2)
ax.set_xlabel('Observation time (s)', fontsize=12)
ax.set_ylabel('Photon count', fontsize=12)
```

Now, what does this pulsar signal look like when it is **added to noise**?

First, let’s look at a **bright pulsar**, where the signal is strong and easy to see.


```python
time, noise = np.load("data/channel_01_telescope_timeseries.npy")
```

```python
bright_pulsar = psr_slow*8
data_faint = noise + np.round(bright_pulsar)
```

```python
fig, (ax1, ax2) = plt.subplots(figsize=(20, 8), nrows=2)

ax1.plot(time, noise, label='noise', c='C0', ls='-', lw=0.5, marker='o', markersize=2)
ax1.plot(time, bright_pulsar, label='pulsar signal', c='C3')
ax1.legend(fontsize=12, loc=1)
ax1.set_xlabel('Observation time (s)', fontsize=12)
ax1.set_ylabel('Photon count', fontsize=12)

ax2.plot(time, data_faint, label='Combined signal', c='C0', ls='-', lw=0.5, marker='o', markersize=2)
ax2.legend(fontsize=12, loc=1)
ax2.set_xlabel('Observation time (s)', fontsize=12)
ax2.set_ylabel('Photon count', fontsize=12)

plt.tight_layout()
```

For a **bright pulsar**, it is still fairly easy to see the signal in the combined time series.

But what happens if the pulsar is **very faint**?

Let’s look at a faint pulsar signal and see how it compares to the noise.


```python
faint_pulsar = psr_slow
data_faint = noise + np.round(faint_pulsar)
```

```python
fig, (ax1, ax2) = plt.subplots(figsize=(20, 8), nrows=2)

ax1.plot(time, noise, label='noise', c='C0', ls='-', lw=0.5, marker='o', markersize=2)
ax1.plot(time, faint_pulsar, label='pulsar signal', c='C3')
ax1.legend(fontsize=12, loc=1)
ax1.set_xlabel('Observation time (s)', fontsize=12)
ax1.set_ylabel('Photon count', fontsize=12)

ax2.plot(time, data_faint, label='Combined signal', c='C0', ls='-', lw=0.5, marker='o', markersize=2)
ax2.legend(fontsize=12, loc=1)
ax2.set_xlabel('Observation time (s)', fontsize=12)
ax2.set_ylabel('Photon count', fontsize=12)

plt.tight_layout()
```

As you can see, for a **faint pulsar** the signal is completely buried in the noise.  
In the combined time series, it is no longer possible to see the pulses by eye.

This does **not** mean the pulsar is gone — it is still there, repeating very regularly.  
The problem is that the signal is much weaker than the noise.

To find faint pulsars like this, astronomers need **more advanced tools**.
